# The Mismatched Test for Markovian Samples

## The Hoeffding and Mismatched Tests


First, let's consider a simple case. Assume we know the PL, denoted $\check{\boldsymbol{\pi}}$, of the alternative hypothesis.

Test statistic for the Hoeffding test:

$$
D^{\text{HT}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = 
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\Gamma _n(\theta_{ij})\big/
	\big( \sum_{t = 1}^N \Gamma _n(\theta_{it})\big)} 
{\pi _{ij}\big/ \big(\sum_{t = 1}^N \pi _{it} \big)}.
$$

Test statistic for the mismatched test:

$$
D^{\text{MM}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = 
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\check{\pi}_{ij}\big/
	\big( \sum_{t = 1}^N \check{\pi}_{it}\big)} 
{\pi _{ij}\big/ \big(\sum_{t = 1}^N \pi _{it} \big)}.
$$

Extension:

Instead of a single PL, we can consider the case where a finite set of PLs, denoted $\check{\boldsymbol{\Pi}}$, is adopted by the alternative hypothesis. In this case, the test statistic for the mismatched test is 

$$
D^{\text{MM}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = \sup_{\check{\boldsymbol{\pi}} \in \check{\boldsymbol{\Pi}}}
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\check{\pi}_{ij}\big/
	\big( \sum_{t = 1}^N \check{\pi}_{it}\big)} 
{\pi _{ij}\big/ \big(\sum_{t = 1}^N \pi _{it} \big)}.
$$

Now, assuming a PL set $\boldsymbol{\Pi}$, rather than a single PL $\boldsymbol{\pi}$, is accepted by the null hypothesis, then the test statistic for the Hoeffding test is

$$
D^{\text{HT}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = 
\inf_{\boldsymbol{\pi} \in \boldsymbol{\Pi}} \sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\Gamma _n(\theta_{ij})\big/
	\big( \sum_{t = 1}^N \Gamma _n(\theta_{it})\big)} 
{\pi _{ij}\big/ \big(\sum_{t = 1}^N \pi _{it} \big)},
$$
and the test statistic for the mismatched test is

$$
D^{\text{MM}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = \inf_{\boldsymbol{\pi} \in \boldsymbol{\Pi}}\sup_{\check{\boldsymbol{\pi}} \in \check{\boldsymbol{\Pi}}}
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\check{\pi}_{ij}\big/
	\big( \sum_{t = 1}^N \check{\pi}_{it}\big)} 
{\pi _{ij}\big/ \big(\sum_{t = 1}^N \pi _{it} \big)}.
$$

## The “Faked” Hoeffding and Mismatched Tests

Test statistic for the "faked" Hoeffding test:

$$
D^{\text{FHT}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = 
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\Gamma _n(\theta_{ij})} 
{\pi _{ij}}.
$$


Test statistic for the "faked" mismatched test:

$$
D^{\text{FMM}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = 
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\check{\pi}_{ij}} 
{\pi _{ij}}.
$$


The "faked" extension:

Instead of a single PL, we can consider the case where a finite set of PLs, denoted $\check{\boldsymbol{\Pi}}$, is accepted by the alternative hypothesis. In this case, the test statistic for the "faked" mismatched test is 

$$
D^{\text{FMM}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = \sup_{\check{\boldsymbol{\pi}} \in \check{\boldsymbol{\Pi}}}
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\check{\pi}_{ij}} 
{\pi _{ij}}.
$$

Now, assuming a PL set $\boldsymbol{\Pi}$, rather than a single PL $\boldsymbol{\pi}$, is accepted by the null hypothesis, then the test statistic for the "faked" Hoeffding test is

$$
D^{\text{FHT}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = 
\inf_{\boldsymbol{\pi} \in \boldsymbol{\Pi}} \sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\Gamma _n(\theta_{ij})} 
{\pi _{ij}},
$$
and the test statistic for the "faked" mismatched test is

$$
D^{\text{FMM}}( {\left. \boldsymbol{\Gamma} _n \right\| \boldsymbol{\pi} }) = \inf_{\boldsymbol{\pi} \in \boldsymbol{\Pi}}\sup_{\check{\boldsymbol{\pi}} \in \check{\boldsymbol{\Pi}}}
\sum_{i = 1}^N \sum_{j = 1}^N \Gamma_n(\theta_{ij}) \log 
\frac{\check{\pi}_{ij}} 
{\pi _{ij}}.
$$

```python
# mu and mu0 are PLs accepted by the null hypothesis
# mu1, mu2, and mu3 are the "anomaly" PLs

def KL_est(x, mu, mu0, mu1, mu2, mu3):
    """
    Estimate the relative entropy (K-L divergence)
    x: a sample path of the chain
    mu: the true stationary distribution; an N x N matrix
    """
    mu = np.array(mu)
    N, _ = mu.shape
    assert(N == _)
    
    mu0 = np.array(mu0)
    N0, _ = mu0.shape
    assert(N0 == _)
    
    mu1 = np.array(mu1)
    N1, _ = mu1.shape
    assert(N1 == _)
    
    mu2 = np.array(mu2)
    N2, _ = mu2.shape
    assert(N2 == _)
    
    mu3 = np.array(mu3)
    N3, _ = mu3.shape
    assert(N3 == _)

    # Compute the empirical distribution
    eps = 1e-8
    gama = [0] * N * N
    for j in range(0, N**2):
        gama[j] = (x.count(j)) / (len(x))
        if gama[j] < eps:
            gama[j] = eps
    for j in range(0, N**2):
        gama[j] = gama[j] / sum(gama)  # Normalize the estimated probability law
    gama = np.array(gama)
    gama = np.reshape(gama, (N, N))

    # Compute the relative entropy (K-L divergence)
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(gama[i, j] / (sum(gama[i, :]))) - log(mu[i, j] / (sum(mu[i, :]))))
            
    KLht = sum(sum(d))

    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(gama[i, j] / (sum(gama[i, :]))) - log(mu0[i, j] / (sum(mu0[i, :]))))
            
    KLht0 = sum(sum(d))
    
    KLht = min(KLht, KLht0)
    
    # Compute the Mismatched divergence
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu1[i, j] / (sum(mu1[i, :]))) - log(mu[i, j] / (sum(mu[i, :]))))
    KLmm1 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu2[i, j] / (sum(mu2[i, :]))) - log(mu[i, j] / (sum(mu[i, :]))))
    KLmm2 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu3[i, j] / (sum(mu3[i, :]))) - log(mu[i, j] / (sum(mu[i, :]))))
    KLmm3 = sum(sum(d))
    
    KLmm = max(KLmm1, KLmm2, KLmm3)
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu1[i, j] / (sum(mu1[i, :]))) - log(mu0[i, j] / (sum(mu0[i, :]))))
    KLmm10 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu2[i, j] / (sum(mu2[i, :]))) - log(mu0[i, j] / (sum(mu0[i, :]))))
    KLmm20 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu3[i, j] / (sum(mu3[i, :]))) - log(mu0[i, j] / (sum(mu0[i, :]))))
    KLmm30 = sum(sum(d))
    
    KLmm0 = max(KLmm10, KLmm20, KLmm30)
    
    KLmm = min(KLmm, KLmm0)

    # Compute the "fake" Hoeffding relative entropy
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(gama[i, j]) - log(mu[i, j]))
    KLfht = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(gama[i, j]) - log(mu0[i, j]))
    KLfht0 = sum(sum(d))
    
    KLfht = min(KLfht, KLfht0)
    
    # Compute the "fake" Mismatched divergence
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu1[i, j]) - log(mu[i, j]))
    KLfmm1 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu2[i, j]) - log(mu[i, j]))
    KLfmm2 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu3[i, j]) - log(mu[i, j]))
    KLfmm3 = sum(sum(d))
    
    KLfmm = max(KLfmm1, KLfmm2, KLfmm3)
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu1[i, j]) - log(mu0[i, j]))
    KLfmm10 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu2[i, j]) - log(mu0[i, j]))
    KLfmm20 = sum(sum(d))
    
    d = np.zeros((N, N))
    for i in range(0, N):
        for j in range(0, N):
            d[i, j] = gama[i, j] * (log(mu3[i, j]) - log(mu0[i, j]))
    KLfmm30 = sum(sum(d))
    
    KLfmm0 = max(KLfmm10, KLfmm20, KLfmm30)
    
    KLfmm = min(KLfmm, KLfmm0)

    return KLht, KLmm, KLfht, KLfmm
```

```python
def wcKLs_testSet(N, n, num_test_sample):
    # mu_0: the initial distribution; 1 x (N**2) 
    # mu: the actual stationary distribution; 1 x (N**2)
    # mu_1: the estimated stationary distribution 
    # P: the new transition matrix (the old transition matrix is Q)
    # G_1: the estimate of the gradient
    # H_1: the estimate of the Hessian

    # P is the ground truth transition matrix (null PL #1)
    mu_0, mu, mu_1, P, G_1, H_1, Sigma_1, W_1, rho_1, Chi_1 = ChainGen(N)  
    
    # PPPPP is another ground truth transition matrix (null PL #2)
    mu_0PPPP, muPPPP, mu_1PPPP, PPPPP, G_1PPPP, H_1PPPP, Sigma_1PPPP, W_1PPPP, rho_1PPPP, Chi_1PPPP = \
    ChainGen(N) 

    # PP is a different transition matrix for testing purposes (anomaly PL #1)
    mu_0P, muP, mu_1P, PP, G_1P, H_1P, Sigma_1P, W_1P, rho_1P, Chi_1P = ChainGen(N) 
    
    # PPP is a further different transition matrix for testing purposes (anomaly PL #2)
    mu_0PP, muPP, mu_1PP, PPP, G_1PP, H_1PP, Sigma_1PP, W_1PP, rho_1PP, Chi_1PP = ChainGen(N) 
    
    # PPPP is a third different transition matrix for testing purposes (anomaly PL #3)
    mu_0PPP, muPPP, mu_1PPP, PPPP, G_1PPP, H_1PPP, Sigma_1PPP, W_1PPP, rho_1PPP, Chi_1PPP = ChainGen(N) 
    
    negative_sample = {}
    positive_sample = {}

    # Get sample paths of the Markov chain with length n; 
    # these paths will be the test set
    for idx in range(int(num_test_sample/4)):
        negative_sample[idx] = chain(mu, P, n)
    for idx in range(int(num_test_sample/4), 2 * int(num_test_sample/4)):
        negative_sample[idx] = chain(mu, PPPPP, n)
    for idx in range(int(num_test_sample/6)):
        positive_sample[idx] = chain(muP, PP, n)
    for idx in range(int(num_test_sample/6), 2 * int(num_test_sample/6)):
        positive_sample[idx] = chain(muPP, PPP, n)
    for idx in range(2 * int(num_test_sample/6), 3 * int(num_test_sample/6)):
        positive_sample[idx] = chain(muPPP, PPPP, n)
    
    mu = np.reshape(mu, (N, N))
    muP = np.reshape(muP, (N, N))
    muPP = np.reshape(muPP, (N, N))
    muPPP = np.reshape(muPPP, (N, N))
    muPPPP = np.reshape(muPPPP, (N, N))
    
    # for Hoeffding
    negative_sample_KL_HT = {}
    positive_sample_KL_HT = {}
    
    # for Mismatched
    negative_sample_KL_MM = {}
    positive_sample_KL_MM = {}
    
    # for "fake" Hoeffding
    negative_sample_KL_FHT = {}
    positive_sample_KL_FHT = {}
    
    # for "fake" Mismatched
    negative_sample_KL_FMM = {}
    positive_sample_KL_FMM = {}
    
    # Compute KL divergence values w.r.t. mu for each sample
    for idx in range(int(num_test_sample/2)):
        negative_sample_KL_HT[idx], negative_sample_KL_MM[idx], negative_sample_KL_FHT[idx], \
        negative_sample_KL_FMM[idx] = KL_est(negative_sample[idx], mu, muPPPP, muP, muPP, muPPP)
        
    for idx in range(int(num_test_sample/2)):
        positive_sample_KL_HT[idx], positive_sample_KL_MM[idx], positive_sample_KL_FHT[idx], \
        positive_sample_KL_FMM[idx] = KL_est(positive_sample[idx], mu, muPPPP, muP, muPP, muPPP)
        
    return negative_sample_KL_HT, positive_sample_KL_HT, negative_sample_KL_MM, positive_sample_KL_MM, \
negative_sample_KL_FHT, positive_sample_KL_FHT, negative_sample_KL_FMM, positive_sample_KL_FMM
```

```python
def roc_data(n, beta, num_test_sample, negative_sample_KL_HT, positive_sample_KL_HT, \
             negative_sample_KL_MM, positive_sample_KL_MM, negative_sample_KL_FHT, positive_sample_KL_FHT, \
            negative_sample_KL_FMM, positive_sample_KL_FMM):
    
    # Get thresholds for Hoeffding's test corresponding to sample length n    
    eta = {}
    key = str(n) + "_" + str(beta)
    eta[key] = beta

    negative_sample_HT = {}
    negative_sample_MM = {}
    negative_sample_FHT = {}
    negative_sample_FMM = {}

    positive_sample_HT = {}
    positive_sample_MM = {}
    positive_sample_FHT = {}
    positive_sample_FMM = {}

    for idx in range(int(num_test_sample/2)):
        if negative_sample_KL_HT[idx] > eta[key]:
            negative_sample_HT[idx] = 1
        else:
            negative_sample_HT[idx] = -1
        if negative_sample_KL_MM[idx] > eta[key]:
            negative_sample_MM[idx] = 1
        else:
            negative_sample_MM[idx] = -1
        if negative_sample_KL_FHT[idx] > eta[key]:
            negative_sample_FHT[idx] = 1
        else:
            negative_sample_FHT[idx] = -1
        if negative_sample_KL_FMM[idx] > eta[key]:
            negative_sample_FMM[idx] = 1
        else:
            negative_sample_FMM[idx] = -1


    for idx in range(int(num_test_sample/2)):
        if positive_sample_KL_HT[idx] > eta[key]:
            positive_sample_HT[idx] = 1
        else:
            positive_sample_HT[idx] = -1
        if positive_sample_KL_MM[idx] > eta[key]:
            positive_sample_MM[idx] = 1
        else:
            positive_sample_MM[idx] = -1
        if positive_sample_KL_FHT[idx] > eta[key]:
            positive_sample_FHT[idx] = 1
        else:
            positive_sample_FHT[idx] = -1
        if positive_sample_KL_FMM[idx] > eta[key]:
            positive_sample_FMM[idx] = 1
        else:
            positive_sample_FMM[idx] = -1
            

    TP_HT = count_one(positive_sample_HT.values())[0]
    FP_HT = count_one(negative_sample_HT.values())[0]
    TN_HT = count_one(negative_sample_HT.values())[1]
    FN_HT = count_one(positive_sample_HT.values())[1]
    TPR_HT = TP_HT / (TP_HT + FN_HT)
    FPR_HT = FP_HT / (FP_HT + TN_HT)

    TP_MM = count_one(positive_sample_MM.values())[0]
    FP_MM = count_one(negative_sample_MM.values())[0]
    TN_MM = count_one(negative_sample_MM.values())[1]
    FN_MM = count_one(positive_sample_MM.values())[1]
    TPR_MM = TP_MM / (TP_MM + FN_MM)
    FPR_MM = FP_MM / (FP_MM + TN_MM)
    
    TP_FHT = count_one(positive_sample_FHT.values())[0]
    FP_FHT = count_one(negative_sample_FHT.values())[0]
    TN_FHT = count_one(negative_sample_FHT.values())[1]
    FN_FHT = count_one(positive_sample_FHT.values())[1]
    TPR_FHT = TP_FHT / (TP_FHT + FN_FHT)
    FPR_FHT = FP_FHT / (FP_FHT + TN_FHT)
    
    TP_FMM = count_one(positive_sample_FMM.values())[0]
    FP_FMM = count_one(negative_sample_FMM.values())[0]
    TN_FMM = count_one(negative_sample_FMM.values())[1]
    FN_FMM = count_one(positive_sample_FMM.values())[1]
    TPR_FMM = TP_FMM / (TP_FMM + FN_FMM)
    FPR_FMM = FP_FMM / (FP_FMM + TN_FMM)
    
    data_HT = {"TPR": TPR_HT, "FPR":FPR_HT}
    data_MM = {"TPR": TPR_MM, "FPR":FPR_MM}
    data_FHT = {"TPR": TPR_FHT, "FPR":FPR_FHT}
    data_FMM = {"TPR": TPR_FMM, "FPR":FPR_FMM}
    
    return data_HT, data_MM, data_FHT, data_FMM
```

```python
N = 4
n = 25
num_test_sample = 12000

beta_list = list(np.arange(-2, 2, 0.0001)[1:-1])
```
